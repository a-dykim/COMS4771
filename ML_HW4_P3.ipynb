```python
def compute_distance(coord1, coord2):
    """
    Given the two coordinate, compute L2 distance
    """
    coord1 = np.array(coord1)
    coord2 = np.array(coord2)
    distance = np.sqrt(np.sum((coord1 - coord2)**2))
    return distance
```

```python
def get_permutation_pairs(length):
    """
    Compute all permutation pairs
    """
    arr = np.arange(length)
    perm = permutations(arr,2)
    seq = []
    for i in list(perm):
        seq.append(i)
   #nseq = len(seq)
    #for 
    
    return seq
```

```python
def compute_distances(coordinates):
    """
    Compute L2 distances between all coordinate combinations
    
    Input: coordinates dictionary
    """
    cities = ['BOS', 'NYC', 'DC', 'MIA', 'CHI', 'SEA', 'SF', 'LA', 'DEN']
    ncity = len(cities)
    pairs = get_permutation_pairs(ncity)
    distance_matrix = np.zeros((ncity, ncity))
    
    for i in range(len(pairs)):
        loc1 = pairs[i][0]
        loc2 = pairs[i][1]
        city1 = np.array(coordinates[cities[loc1]])
        city2 = np.array(coordinates[cities[loc2]])
        distance_matrix[loc1, loc2] = np.sqrt(np.sum((city1 - city2)**2))
    
    return np.array(distance_matrix)

def compute_difference(coordinates):
    """
    Compute the differnce between all coordinates
    """
    cities = ['BOS', 'NYC', 'DC', 'MIA', 'CHI', 'SEA', 'SF', 'LA', 'DEN']
    ncity = len(cities)
    pairs = get_permutation_pairs(ncity)
    diff = {}
    diff_matrix = np.zeros((ncity, ncity,2))
    
    for i in range(len(pairs)):
        loc1 = pairs[i][0]
        loc2 = pairs[i][1]
        city1 = np.array(coordinates[cities[loc1]])
        city2 = np.array(coordinates[cities[loc2]])
        #print (city1 - city2)
        diff_matrix[loc1, loc2] = city1 - city2
    
    return diff_matrix
```

```python
from itertools import permutations 
def unique_sequences(length):
    arr = np.arange(length)
    perm = permutations(arr)
    seq = []
    for i in list(perm):
        seq.append(i)
    return seq
```

```python
def rand_draw_noverlap(avoid, low, high):
    """
    Draw a random number avoiding a number
    """
    rand = np.random.randint(low, high, 1)
    while rand is avoid:
        rand = np.random.randint(low, high, 1)
    return int(rand)
        
        
```
